```python
# Generate embeddings using sentence-transformers
# pip install sentence-transformers

from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
    "I love data engineering",
    "I enjoy building data pipelines", # Similar meaning
    "The cat sat on the mat"  # Different meaning
]

embeddings = model.encode(sentences)

print(embeddings)
print(embeddings.shape)

# Check similarity; closer to 1.0 = more similar
def cosine_sim(a,b):
  return np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))

sim1 = cosine_sim(embeddings[0], embeddings[1])
sim2 = cosine_sim(embeddings[0], embeddings[2])

# Check the similarity of the sentences ,
# Inference: If the values are closer to 1.0  " Very similar & far from 1.0 : Very different
print(f" 'data engineering' vs 'data pipelines' : {sim1:.3f}")
print(f" 'data engineering' vs 'cat on mat' : {sim2:.3f}")

```
