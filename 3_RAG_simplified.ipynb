```python
# A MINIMAL RAG in ~20 lines - just to see the full picture
# I will build this properly in phase 3

# 0. import required libraries
from sentence_transformers import SentenceTransformer
import numpy as np

# 1. define the functions
def cosine_sim(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

#2. Assign variables
model = SentenceTransformer('all-MiniLM-L6-v2')
# Your "knowledge base" (later this becomes a vector DB)
documents = [
    "Spark is used for distributed data processing",
    "AWS Bedrock provides managed LLM services",
    "Data Governance ensures data quality and compliance",
    "Pinecone is managed vector database"
]

# Step 4: Embed all documents (do this once , store in DB)
doc_embeddings = model.encode(documents)

# Step 5: User asks a question; Use different queries for testing
#query = "What took helps with large scale data processing?"
#query = "What is the named of managed database"
query = "Define llm services"
query_embedding = model.encode([query])[0]

# Step 6:Find most similar document (retreival!)
# we use context below for ease of reading & computation
similarities = [cosine_sim(query_embedding,d) for d in doc_embeddings]

best_match_idx = np.argmax(similarities)

# 7. Display the results
print(f"Query: {query}")
print(f"Retreived: {documents[best_match_idx]}")
```
