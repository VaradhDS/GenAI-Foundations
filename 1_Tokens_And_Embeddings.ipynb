```python
# This is how GPT tokenizes text
# pip install tiktoken

#0. import & assign variable
import tiktoken
encoder = tiktoken.encoding_for_model('gpt-4')

# 1. Assignment & tokenization
text = "I love data engineering and machine learning"
tokens = encoder.encode(text)
token_words = [encoder.decode([t])
for t in tokens]

print("---------------------")
print(f"Token ID's: {tokens}")
print("---------------------")
print(f"Token Words: {token_words}")
print("---------------------")
print(f"Token count: {len(tokens)}")
print("---------------------")
```
